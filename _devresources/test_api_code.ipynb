```python
import requests 
import json
import pandas as pd 
import numpy as np 
```

# Test the coins/markets endpoint 

```python
def get_endpoint(endpoint, *args):
    
    for arg in args:
        endpoint = endpoint + arg
    
    return endpoint

base_url = 'https://api.coingecko.com/api/v3/'
query_params = 'coins/markets?vs_currency=usd&order=market_cap_des&per_page=100&page=1'

# get the coin markets endpoint
query_url = get_endpoint(base_url, query_params)
```

```python
def make_request(url):
    try:
        response = requests.get(url)
        # Check if the request was successful (status code 200)
        response.raise_for_status()
        return response
    except requests.exceptions.RequestException as e:
        # Handle any request exception (e.g., connection error, timeout, etc.)
        print(f"Error making the request: {e}")
        return None

response = make_request(query_url)
```

```python
test = response.json()
```

```python
for i in test:
    print(i)
```
